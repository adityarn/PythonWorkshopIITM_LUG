```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatch
import numpy as np
```

```python
x = np.linspace(0, 10, 101)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([1.2, 0, 1, 1])
ax3 = fig.add_axes([0.6, -1.2, 1, 1])
#---
ax1.plot(x, y1, "b.-")

ax1.set_xlim([-1, 11])
ax1.grid()

ax1.set_xlabel("x")
ax1.set_ylabel("sin(x)")
#---
ax2.plot(y1, y2, "g*")
ax2.title.set_text("cos(x) vs sin(x)")
#---
ax3.plot(x, y1, "ro", label="sine")

ax3.axvline(x=np.pi, linewidth=1, color="c", ymin=0.5)
ax3.axvline(x=2*np.pi, linewidth=4, color="k", ymin=0.25, ymax=1)
ax3.axhline(y=0, linewidth=1, color="m")

ax3.set_ylim([-2, 1])

MyArrow = mpatch.Arrow(x=np.pi, dx=0, y=-1, dy=0.45, width=0.5, color="c")
ax3.add_patch(MyArrow)

MyText = "This is x=$\pi$"
ax3.text(x=np.pi, y=-1.1, s=MyText, horizontalalignment="center")

ax3.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)
#---
plt.show()
```
